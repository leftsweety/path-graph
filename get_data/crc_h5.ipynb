```python
import pandas as pd

h5_label_path = './data_sheets/CRC_CLI_M21-404-HE_harmonized_feature_table.xlsx'
file_path_crc = './data_sheets/FINAL_Reconciled_sampleID_clinical_CRC.xlsx'
image_id_res = './data_sheets/image_id_res.csv'

df_crc = pd.read_excel(file_path_crc, sheet_name='Outcome Statistical (Internal)', na_values=['NA', 'N/A'])
df_crc=df_crc.dropna(subset = ['HE Concentriq Image Storage Key'])
crc_hae_label_df = pd.read_excel(h5_label_path, na_values=['NA', 'N/A'])
```

```python
df_crc=df_crc.iloc[:, :2]
df_crc['TDx Number:\n[TDx_Number]']
crc_hae_label_df=crc_hae_label_df.iloc[:, :2]
```

```python

merged_df = pd.merge(df_crc, crc_hae_label_df, left_on='TDx Number:\n[TDx_Number]', right_on='CASE_ID', how='inner')
```

```python
image_id_res_df = pd.read_csv(image_id_res)
HE_id = []
for row in image_id_res_df.iterrows():
    row=row[1]
    if row['cancer']=='CRC':
        subj_id = row.subj_id
        HE_id.append(list(merged_df[merged_df['Subjects']==subj_id]['H & E_ID'])[0])
    else:
        HE_id.append('NA')
image_id_res_df['HE_id']=HE_id
```

```python
image_id_res_df.to_csv(image_id_res)
```
